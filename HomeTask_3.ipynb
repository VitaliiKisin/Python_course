#### 1.1 Создать новый признак price_per_sq_lot, который будет содержать среднюю стоимость за один кв. метр общей площади

```python
df['sqft_lot'] = pd.to_numeric(df['sqft_lot'], errors='coerce') # Преобразовать данные в столбцах в числовые
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
print('Средняя цена sqft_lot :\n', df['price_per_sq_lot'])
```

```python
df['sqft_lot'] = pd.to_numeric(df['sqft_lot'], errors='coerce')
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price_per_sq_lot'] = df['price'] / df['sqft_lot']
print('Средняя цена первых десяти домов:\n', df['price_per_sq_lot'].head(10))
```
