#### 1.2 Создать новый признак Memory_Amount, который будет содержать только количество Gb памяти без указания типа носителя
#### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)
#### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.head()
```

```python
df.Memory.value_counts()
```

```python
def gigs(text) :
    result = text.split()[0]
    if 'GB' in result:
        result = result.replace('GB','')
    elif 'TB' in result:
        result = float(result.replace('TB',''))*1024
        return int(result)
```

```python
df['Memory_Amount'] = df['Memory'].apply(lambda x: gigs(x))
df['Memory_Amount'].value_counts
```

#### 1.3 Создать новый признак Memory_Type, который будет содержать только тип носителя (HDD/SDD/др.)


```python
'128GB SSD +  1TB HDD '.split('B ')[-1]
```

```python
df['Memory_Type'] = df['Memory'].apply(lambda x: x.split('B ')[-1])
df['Memory_Type'].value_counts()
```

#### 1.4 Удалите признаки Memory и ScreenResolution

```python
df.drop(['Memory', 'ScreenResolution'], axis = 1, inplace = True)
```
